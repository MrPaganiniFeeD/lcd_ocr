```python
from ultralytics import YOLO
from model import ResNetPredictor

def detection_temperature(image_path: str, model_path: str) -> float:
    model = YOLO(model_path)
    results = model.predict(image_path)
    
    result = results[0]
    print(f"Обнаружено объектов: {len(result.boxes)}")

    out_detection = []
    for i, box in enumerate(result.boxes):
        class_id = int(box.cls[0])
        confidence = float(box.conf[0])
        cls_name = result.names[class_id]
        bbox = box.xyxy[0].tolist()  # [x1, y1, x2, y2]
        x1 = bbox[0]
        x2 = bbox[2]
        out_detection.append([(x1 + x2) / 2, cls_name])
        print(f"Объект {i+1}:")
        print(f"  Класс: {class_id} ({result.names[class_id]})")
        print(f"  Уверенность: {confidence:.3f}")
        print(f"  BBOX: {bbox}")
    

    out_detection.sort()
    out_value = ''
    is_node = False

    for i in range(len(out_detection)):
        name = out_detection[i][1]

        if name == '.' and is_node:
            pass

        if name == '.' and is_node == False:
            is_node = True

        out_value += out_detection[i][1]
    
    try:
        out_value = float(out_value)
    except ValueError as e:
        print("Мы неправильно считали температуру")
    print("Температура:", out_value)
    result.show()
    return out_value
```
